# **Check Missing Values**

```python
# Menghitung jumlah nilai yang hilang (missing values) di setiap kolom
missing_values = df.isnull().sum()

# Menampilkan jumlah nilai yang hilang per kolom
print("Jumlah nilai yang hilang per kolom:")
print(missing_values)
```

```python
# Menampilkan jumlah missing values sebelum imputasi
print("Jumlah nilai yang hilang per kolom sebelum imputasi:")
print(df.isnull().sum())
```

```python
# Mengisi missing values pada kolom numerik dengan median
df['Premis Cd'] = df['Premis Cd'].fillna(df['Premis Cd'].median())  # Mengisi dengan median untuk kolom numerik

# Mengisi missing values pada kolom kategorikal dengan modus
df['Vict Sex'] = df['Vict Sex'].fillna(df['Vict Sex'].mode()[0])  # Mengisi dengan modus untuk kolom kategorikal
df['Vict Descent'] = df['Vict Descent'].fillna(df['Vict Descent'].mode()[0])  # Mengisi dengan modus untuk kolom kategorikal
df['Crm Cd 1'] = df['Crm Cd 1'].fillna(df['Crm Cd 1'].mode()[0])  # Mengisi dengan modus untuk kolom kategorikal

# Mengisi missing values pada kolom numerik menggunakan interpolasi
df['LAT'] = df['LAT'].interpolate()  # Mengisi dengan interpolasi untuk kolom numerik
df['LON'] = df['LON'].interpolate()  # Mengisi dengan interpolasi untuk kolom numerik

# Mengisi missing values pada kolom kategorikal dengan nilai tetap
df['Weapon Used Cd'] = df['Weapon Used Cd'].fillna('Unknown')  # Mengisi dengan nilai tetap untuk kolom kategorikal
```

```python
# Mengisi missing values dengan modus atau nilai tetap
df['Premis Desc'] = df['Premis Desc'].fillna(df['Premis Desc'].mode()[0])
df['Weapon Desc'] = df['Weapon Desc'].fillna(df['Weapon Desc'].mode()[0])
df['Crm Cd 2'] = df['Crm Cd 2'].fillna(df['Crm Cd 2'].mode()[0])
df['Crm Cd 3'] = df['Crm Cd 3'].fillna(df['Crm Cd 3'].mode()[0])
df['Crm Cd 4'] = df['Crm Cd 4'].fillna(df['Crm Cd 4'].mode()[0])
df['Cross Street'] = df['Cross Street'].fillna(df['Cross Street'].mode()[0])
```

```python
# Kolom yang dianggap sebagai data numerik (jika memang numerik)
numeric_columns = ['Mocodes', 'Premis Desc', 'Weapon Desc', 'Crm Cd 2', 'Crm Cd 3', 'Crm Cd 4', 'Cross Street']

# Kolom yang dianggap sebagai data kategorikal
categorical_columns = ['Vict Sex', 'Vict Descent', 'Weapon Used Cd', 'Crm Cd 1']
```

```python
# Mengisi missing values pada kolom kategorikal dengan modus
for col in categorical_columns:
    df[col] = df[col].fillna(df[col].mode()[0])

# Mengisi missing values pada kolom kategorikal dengan nilai tetap (jika ada)
df['Weapon Used Cd'] = df['Weapon Used Cd'].fillna('Unknown')
```

```python
# Menampilkan jumlah nilai yang hilang per kolom setelah imputasi
missing_values_after_additional = df.isnull().sum()
print("Jumlah nilai yang hilang per kolom setelah imputasi tambahan:")
print(missing_values_after_additional)
```

```python
# Mengisi missing value di kolom 'Status' dengan nilai yang sesuai (misalnya modus atau nilai tetap)
df['Status'] = df['Status'].fillna(df['Status'].mode()[0])
```

```python
# Cek tipe data kolom 'Mocodes'
print(df['Mocodes'].dtype)

# Jika 'Mocodes' bukan numerik, coba ubah atau tangani data
df['Mocodes'] = pd.to_numeric(df['Mocodes'], errors='coerce')  # Ubah menjadi numerik, set data yang tidak valid menjadi NaN

# Imputasi ulang dengan median setelah memastikan tipe data
df['Mocodes'] = df['Mocodes'].fillna(df['Mocodes'].median())
```

```python
# Menampilkan jumlah nilai yang hilang per kolom setelah imputasi tambahan
missing_values_after_final = df.isnull().sum()
print("Jumlah nilai yang hilang per kolom setelah imputasi final:")
print(missing_values_after_final)
```

```python
# Imputasi dengan waktu default, misalnya 00:00:00
df['TIME OCC'].fillna(pd.to_datetime('00:00:00').time(), inplace=True)
```

```python
# Mengisi dengan mode waktu
mode_time = df['TIME OCC'].mode()[0]
df['TIME OCC'].fillna(mode_time, inplace=True)
```

```python
# Mengubah kolom ke format datetime dan mengisi missing values dengan interpolasi
df['TIME OCC'] = pd.to_datetime(df['TIME OCC'], format='%H:%M:%S', errors='coerce')
df['TIME OCC'] = df['TIME OCC'].interpolate(method='linear')
df['TIME OCC'] = df['TIME OCC'].dt.time
```
